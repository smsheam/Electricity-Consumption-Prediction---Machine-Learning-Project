Import necessary libraries

```python
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
import seaborn as sns
```

Load data

```python
df = pd.read_csv('/content/welding_cost_final.csv')
```

Train models

```python
model1 = linear_model.LinearRegression()
model1.fit(df[['length', 'breadth', 'depth', 'ship_type']], df['filler_metal'])
model2 = linear_model.LinearRegression()
model2.fit(df[['length', 'breadth', 'depth', 'ship_type']], df['electricity_consumption'])
```

Make predictions

```python
df['filler_metal_pred'] = model1.predict(df[['length', 'breadth', 'depth', 'ship_type']])
df['electricity_consumption_pred'] = model2.predict(df[['length', 'breadth', 'depth', 'ship_type']])
```

Scatter plot for actual vs predicted filler metal consumption

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['filler_metal'], df['filler_metal_pred'], color='blue')
plt.plot([df['filler_metal'].min(), df['filler_metal'].max()],
[df['filler_metal'].min(), df['filler_metal'].max()], 'k--', lw=2)
plt.xlabel('Calculated Electrode Consumption')
plt.ylabel('Predicted Electrode Consumption')
plt.title('Calculated vs Predicted Electrode Consumption')
plt.show()
```

Scatter plot for actual vs predicted electricity consumption

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['electricity_consumption'], df['electricity_consumption_pred'], color='green')
plt.plot([df['electricity_consumption'].min(), df['electricity_consumption'].max()],
[df['electricity_consumption'].min(), df['electricity_consumption'].max()], 'k--', lw=2)
plt.xlabel('Calculated Electricity Consumption')
plt.ylabel('Predicted Electricity Consumption')
plt.title('Calculated vs Predicted Electricity Consumption')
plt.show()
```

Heatmap for correlation matrix

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
